```python
#Loading second data set
df2= pd.read_csv("bse_dataset2.csv")
#Printing the df2
df2
```

```python
#Checking for null values in second df2
df2.isnull().sum().sort_values(ascending=False)
```

```python
#Checking for data types
df2.dtypes
```

```python
#Check for null values in percentagies
for i in df2.columns:
    null_rate = df2[i].isnull().sum() / len(df2) * 100
    if null_rate > 0:
        print("{} null rate: {}%".format(i, round(null_rate, 2)))
```

```python
# Check null values in percentiles
for i in df2.columns:
    if df2[i].isnull().any():
        print(i,"..........",df2[i].isnull().sum() * 100 / df2.shape[0])
```

```python
#Checking null values in percentage
import missingno as msno
msno.bar(df2, color='gray', figsize=(10, 5))
```
